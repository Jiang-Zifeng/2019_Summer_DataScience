```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.neighbors import KNeighborsRegressor
import statsmodels.api as sm
from statsmodels.api import OLS
%matplotlib inline
```

```python
def openTrainData():  #Read train data and storage it in two lists:
    with open('train.txt','r') as f:
        dataset = f.readlines()
    data_dict = {}
    for datalines in dataset:
        data_splits = datalines.strip('\n').split(',')
        data_dict[data_splits[0]] = data_splits[1]
    keys = list(data_dict.keys())
    values = list(data_dict.values())
    return keys,values
```

```python
def plotMaking(x,y):  #Making scatter plot:
    plt.scatter(x,y)
    plt.savefig('train.png', bbox_inches='tight')
    plt.xlabel(input("Please input the x label: "))
    plt.ylabel(input("Please input the y label: "))
    plt.title(input("Please input the title of the plot: "))
    plt.show()
```

```python
def main():  #Main
    np.set_printoptions(suppress=True)
    TimeMin,PickupCount = openTrainData()
    x = []
    y = []
    for i in TimeMin:
        i = float(i)
        x.append(i)
    for j in PickupCount:
        j = float(j)
        y.append(j)
    plotMaking(x,y)
```

```python
if __name__ == '__main__':
    main()
```
